# Problem 1:

Plot Sine and Cosine Graph. 

Then show Cos(0) and Sin(150) degree value using Annotate. 

The plot graph should look like this.

image.png

```python
from matplotlib.patches import ConnectionStyle
# Code Here
import numpy as np
import matplotlib.pyplot as plt

a= np.arange(-2*np.pi, 2*np.pi, 0.01)
x=np.cos(a)
y=np.sin(a)
plt.plot(a, x, color='red')
plt.plot(a, y, color='red')

plt.xticks([-6.28,-4.71,-3.14,-1.57,0,1.57,3.14,4.17,6.28],
           ['$-2\pi$',r'$-3\frac{\pi}{2}$','$-\pi$',r'$-\frac{\pi}{2}$','$0$',r'$\frac{\pi}{2}$','$\pi$',r'$3\frac{\pi}{2}$', '$2\pi$'])

plt.yticks([-1,0,1],
           ['$-1$',r'$0$','$+1$'])

plt.plot([0,0],[-1,1], color='blue', linestyle='--')
plt.plot([2.62,2.62], [-1,0.5], color='blue', linestyle='--')

plt.annotate(r'$cos(0)=1$', xy=(0,1), xytext = (+10,-50),textcoords='offset points',
             fontsize = 10,arrowprops=dict(arrowstyle='->', connectionstyle="arc3, rad=.5"))
plt.annotate(r'sin($\frac{5\pi}{6}=\frac{1}{2}$', xy=(np.pi*5/6,0.5),
              textcoords='offset points', fontsize = 10, arrowprops= dict(arrowstyle='->', connectionstyle="arc3, rad=.5"))
              
```
