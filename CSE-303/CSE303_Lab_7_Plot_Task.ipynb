# Annotate 
Annotate the point xy with text text.

In the simplest form, the text is placed at xy.

Optionally, the text can be displayed in another position xytext. An arrow pointing from the text to the annotated point xy can then be added by defining arrowprops.

\

Learn More:

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot
plt.plot([1, 2, 3, 4])

# Label
plt.ylabel('Some Numbers')

# Set x limits
plt.xlim(0,3.0)

# Set y limits
plt.ylim(1.0,4.0)

# Set x ticks
plt.xticks([0,1,2,3],
           ['$0$', '$1$', '$2$','$3$'])

# plt.xticks([0,1,2,3],
#            ['$-\pi$', r'$-\frac{\pi}{2}$', '$0$', '$+\pi/2$'])

# Using mathtext to get Unicode Characters by wrapping the format specifier with $

# Set y ticks
plt.yticks([1,2,3,4],
           ['$1$', '$2$', '$3$','$4$'])


# Dotted line plot
plt.plot([2, 2], [1, 3], color = 'red', linestyle = "--", linewidth = 1.0)

# Annotate
plt.annotate(r'$y = 3$', xy = (2, 3), xytext = (+50, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))

# Title
plt.title("Line")

plt.show()
```

```python
# Code Here
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi,100)
plt.plot(x, np.cos(x), '-',color='red')

plt.xticks([-6.28,-4.71,-3.14,-1.57,0,1.57,3.14,4.71,6.28],
          ['$-2\pi$',r'$-3\frac{\pi}{2}$','$-\pi$',r'$-\frac{\pi}{2}$','$0$',r'$\frac{\pi}{2}$', '$\pi$', r'$3\frac{\pi}{2}$', '$2\pi$'])

plt.yticks([-1,0,1],
          ['$-1$',r'$0$', '$+1$'])

plt.plot([np.pi/3,np.pi/3], [-1,0.5], color = 'red', linestyle = "--", linewidth = 1.0)
plt.annotate(r'$cos(-\frac{\pi}{3})=1/2$', xy=(np.pi/3,0.5), xytext = (+10, -50),
             textcoords = 'offset points', fontsize = 10, 
             arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))
```

# Exercise 2:

Draw a box plot from "adult.data" with age column and grouping by sex column.

Use df.boxplot() for plotting. 

Use Parameters:

Grid should be false.

Also, give appropriate figure size.

Now use Annotate to pinpoint:


*   Median (With Value)
*   Upper Quartile (Q3) (With Value)
*   Lower Quartile (Q1) (With Value)
*   Upper Whisker (With Value)
*   Lower Whisker (With Value)

Also, pinpoint one outlier (If any).

 


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/adult.data', header=None)
df.columns = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain',
              'capital-loss','hours-per-week','native-country', 'salary']
df.head(10)

# Code Here

ax = df.boxplot(by="sex", column="age", figsize=(10, 8))


mdn = df.groupby('sex')['age'].median()
plt.annotate(r'$Median=%.2f$'%mdn[0], xy = (1,mdn[0]), xytext = (+20, -20), textcoords = 'offset points',
             fontsize = 12, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))
plt.annotate(r'$Median=%.2f$'%mdn[1], xy = (2,mdn[1]), xytext = (+20, -20), textcoords = 'offset points',
             fontsize = 12, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))


q1 = df.groupby('sex')['age'].quantile(.25)
plt.annotate(r'$Lower Quartile(Q3)=%.2f$'%q1[0], xy = (1,q1[0]), xytext = (+20, -20), textcoords = 'offset points',
             fontsize = 12, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))
plt.annotate(r'$Lower Quartile(Q3)=%.2f$'%q1[1], xy = (2,q1[1]), xytext = (+20, -20), textcoords = 'offset points',
             fontsize = 12, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))

q3 = df.groupby('sex')['age'].quantile(.75)
plt.annotate(r'$Upper Quartile(Q3)=%.2f$'%q3[0], xy = (1,q3[0]), xytext = (+20, -20), textcoords = 'offset points',
             fontsize = 12, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))
plt.annotate(r'$Upper Quartile(Q3)=%.2f$'%q3[1], xy = (2,q3[1]), xytext = (+20, -20), textcoords = 'offset points',
             fontsize = 12, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .5"))

```
